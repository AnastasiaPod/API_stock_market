Для работы с API Yahoo Finance будем использовать библиотеку yfinance.

```python
# Далее, импортируем нужные библиотеки для работы

import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Загрузка данных по символу акции AAPL
data = yf.download('AAPL', start='2021-01-01', end='2021-12-31')

# Построение графика линии, отображающего цену закрытия акций
plt.plot(data['Close'])
plt.title('Цена закрытия акций AAPL в 2021 году')
plt.xlabel('Дата')
plt.ylabel('Цена закрытия, $')
plt.show()
```

С использованием данного кода мы загружаем исторические данные о ценах акций компании Apple (AAPL) с помощью библиотеки yfinance и строим график линии, отображающий цену закрытия акций в течение 2021 года.
